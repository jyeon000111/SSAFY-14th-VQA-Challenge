```python
# [셀 1] 공통 경로/설정
DATA_DIR = "/content"  # 이미 /content/train.csv가 있다면 그대로 사용
# 혹시 데이터가 Drive에 있다면 아래로 바꿔도 됨: DATA_DIR = "/content/drive/MyDrive/aichallenge"

TRAIN_CSV = f"{DATA_DIR}/train.csv"
PILOT_CSV = "/content/drive/MyDrive/aichallenge/pilot_300.csv"  # 파일럿은 Drive에 고정 저장
RANDOM_SEED = 42
PILOT_SAMPLES = 300
```

```python

# [셀 2] 파일럿 300개를 pilot_300.csv로 저장 (이미 있으면 덮어쓰기 OK)
import pandas as pd, random, os

assert os.path.exists(TRAIN_CSV), f"TRAIN_CSV가 없습니다: {TRAIN_CSV}"
train_df = pd.read_csv(TRAIN_CSV)

random.seed(RANDOM_SEED)
idxs = list(range(len(train_df)))
random.shuffle(idxs)
pilot_df = train_df.iloc[idxs[:min(PILOT_SAMPLES, len(train_df))]].reset_index(drop=True)

os.makedirs(os.path.dirname(PILOT_CSV), exist_ok=True)
pilot_df.to_csv(PILOT_CSV, index=False)
print(f"✅ 파일럿 데이터 {len(pilot_df)}개 저장 완료 → {PILOT_CSV}")
pilot_df.head(2)
```
