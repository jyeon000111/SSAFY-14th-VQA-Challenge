```python
# âœ… Qwen2.5-VL-7B-Instruct ì œë¡œìƒ·: ëª¨ë¸ ë¡œë“œ + 3ìƒ˜í”Œ ìŠ¤ëª¨í¬ í…ŒìŠ¤íŠ¸ (í•œê¸€ ì£¼ì„ + ì‹œê°„ ì¸¡ì •)
import time, torch
from PIL import Image
import pandas as pd
from transformers import AutoModelForVision2Seq, AutoProcessor, BitsAndBytesConfig

device = "cuda" if torch.cuda.is_available() else "cpu"
print("ë””ë°”ì´ìŠ¤:", device)

# ì´ë¯¸ì§€ í•´ìƒë„: T4ì—ì„œ 384 ê¶Œì¥ (í¬ë©´ ê·¹ë‹¨ì ìœ¼ë¡œ ëŠë ¤ì§)
IMAGE_SIZE = 384
MODEL_ID = "Qwen/Qwen2.5-VL-7B-Instruct"

# 4bit ì–‘ìí™” ì„¤ì • (T4 ì•ˆì • ì‹¤í–‰)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)

t0 = time.perf_counter()
processor = AutoProcessor.from_pretrained(
    MODEL_ID,
    min_pixels=IMAGE_SIZE*IMAGE_SIZE,
    max_pixels=IMAGE_SIZE*IMAGE_SIZE,
    trust_remote_code=True,
)
model = AutoModelForVision2Seq.from_pretrained(
    MODEL_ID,
    quantization_config=bnb_config,
    device_map="auto",            # GPUê°€ ìˆìœ¼ë©´ ìë™ìœ¼ë¡œ cudaë¡œ ì˜¬ë¼ê°
    trust_remote_code=True,
)
model.eval()
if torch.cuda.is_available(): torch.cuda.synchronize()
print(f"[ëª¨ë¸ ë¡œë“œ] {(time.perf_counter()-t0):.2f}s")

SYSTEM_INSTRUCT = (
    "You are a helpful visual question answering assistant. "
    "Answer using exactly one letter among a, b, c, or d. No explanation."
)

def build_mc_prompt(question, a, b, c, d):
    return (
        f"{question}\n"
        f"(a) {a}\n(b) {b}\n(c) {c}\n(d) {d}\n\n"
        "ì •ë‹µì„ ë°˜ë“œì‹œ a, b, c, d ì¤‘ í•˜ë‚˜ì˜ ì†Œë¬¸ì í•œ ê¸€ìë¡œë§Œ ì¶œë ¥í•˜ì„¸ìš”."
    )

def extract_choice(text: str) -> str:
    text = text.strip().lower()
    lines = [l.strip() for l in text.splitlines() if l.strip()]
    if not lines: return "a"
    last = lines[-1]
    if last in ["a","b","c","d"]: return last
    for tok in last.split():
        if tok in ["a","b","c","d"]:
            return tok
    return "a"

# ğŸ” ìŠ¤ëª¨í¬ í…ŒìŠ¤íŠ¸: ì• 3ê°œ ìƒ˜í”Œë§Œ ëŒë ¤ì„œ ì†ë„/ì •ìƒ ë™ì‘ í™•ì¸
TEST_CSV = "/content/test.csv"
test_df = pd.read_csv(TEST_CSV)
assert {"id","path","question","a","b","c","d"}.issubset(test_df.columns), "test.csv ì»¬ëŸ¼ í™•ì¸ í•„ìš”"

n_smoke = min(3, len(test_df))
print(f"ìŠ¤ëª¨í¬ í…ŒìŠ¤íŠ¸ ìƒ˜í”Œ ìˆ˜: {n_smoke}")

t_all = time.perf_counter()
for i in range(n_smoke):
    row = test_df.iloc[i]
    img = Image.open(row["path"]).convert("RGB")

    messages = [
        {"role": "system", "content": [{"type": "text", "text": SYSTEM_INSTRUCT}]},
        {"role": "user", "content": [
            {"type": "image", "image": img},
            {"type": "text",  "text": build_mc_prompt(row["question"], row["a"], row["b"], row["c"], row["d"])},
        ]},
    ]
    text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = processor(text=[text], images=[img], return_tensors="pt").to(device)

    t1 = time.perf_counter()
    with torch.no_grad():
        # temperature ê°™ì€ ë¶ˆí•„ìš” í”Œë˜ê·¸ ì œê±°, max_new_tokens=2ë¡œ ì§§ê²Œ
        out_ids = model.generate(
            **inputs,
            max_new_tokens=2,
            do_sample=False,
            eos_token_id=processor.tokenizer.eos_token_id,
        )
    if torch.cuda.is_available(): torch.cuda.synchronize()
    dt = time.perf_counter() - t1

    output_text = processor.batch_decode(out_ids, skip_special_tokens=True)[0]
    print(f"[{i}] pred={extract_choice(output_text)} | {dt:.2f}s")

t_total = time.perf_counter() - t_all
if torch.cuda.is_available(): torch.cuda.synchronize()
print(f"[ìŠ¤ëª¨í¬ ì´ ì†Œìš”] {t_total:.2f}s | ìƒ˜í”Œë‹¹ í‰ê·  {t_total/n_smoke:.2f}s")
```

```python
# âœ… ì „ì²´ ì¶”ë¡  ì‹¤í–‰ + ì‹œê°„ ì¸¡ì • + ì œì¶œ íŒŒì¼ ì €ì¥
import time, pandas as pd
from PIL import Image
from tqdm import tqdm

TEST_CSV = "/content/test.csv"
SAVE_PATH = "/content/submission2_5_7b.csv"

test_df = pd.read_csv(TEST_CSV)
preds = []

t0 = time.perf_counter()
for i in tqdm(range(len(test_df)), desc="Zero-shot Inference", unit="sample"):
    row = test_df.iloc[i]
    img = Image.open(row["path"]).convert("RGB")

    messages = [
        {"role": "system", "content": [{"type": "text", "text": SYSTEM_INSTRUCT}]},
        {"role": "user", "content": [
            {"type": "image", "image": img},
            {"type": "text",  "text": build_mc_prompt(row["question"], row["a"], row["b"], row["c"], row["d"])},
        ]},
    ]
    text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = processor(text=[text], images=[img], return_tensors="pt").to(device)

    with torch.no_grad():
        out_ids = model.generate(
            **inputs,
            max_new_tokens=2,
            do_sample=False,
            eos_token_id=processor.tokenizer.eos_token_id,
        )
    output_text = processor.batch_decode(out_ids, skip_special_tokens=True)[0]

    # 1. ì¶”ë¡  ê²°ê³¼ë¥¼ ë³€ìˆ˜ì— ë¨¼ì € ì €ì¥
    final_answer = extract_choice(output_text)
    preds.append(final_answer)
    # 2. âœ… ì—¬ê¸°ì— ì¡°ê±´ë¬¸ ì¶”ê°€
    if i < 5:
        print(f"[{i}ë²ˆì§¸ ìƒ˜í”Œ] Raw: '{output_text}'  ->  ì¶”ì¶œëœ ë‹µ: '{final_answer}'")

# â±ï¸ ì´/í‰ê·  ì‹œê°„ ì¶œë ¥
if torch.cuda.is_available(): torch.cuda.synchronize()
elapsed = time.perf_counter() - t0
print(f"[ì¶”ë¡ ] ì´ {elapsed/60:.2f}ë¶„, ìƒ˜í”Œë‹¹ {elapsed/len(test_df):.2f}ì´ˆ")

submission = pd.DataFrame({"id": test_df["id"], "answer": preds})
submission.to_csv(SAVE_PATH, index=False)
print("Saved:", SAVE_PATH)

# ìƒìœ„ 5í–‰ í™•ì¸
submission.head()
```
